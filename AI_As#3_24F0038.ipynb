```python
import numpy as np
import matplotlib.pyplot as plt
import random


def f(x):
    """f(x) = -(x-3)^2 + 9"""
    return -(x - 3)**2 + 9


def hill_climbing(x0=0.0, step_size=0.1, method='simple'):

    def _run_single(x_start, method_type, show_output=True):
        """Run a single hill climbing instance"""
        x = x_start
        path = [x]
        iteration = 0

        if show_output:
            print(f"Starting at x = {x:.4f}, f(x) = {f(x):.4f}")

        while True:
            current_value = f(x)

            if method_type == 'simple':
                #move to first improving neighbor
                neighbor_plus = x + step_size
                if f(neighbor_plus) > current_value:
                    x = neighbor_plus
                    path.append(x)
                    if show_output:
                        print(f"Iteration {iteration + 1}: Moved to x = {x:.4f}, f(x) = {f(x):.4f} (forward)")
                    iteration += 1
                    continue

                neighbor_minus = x - step_size
                if f(neighbor_minus) > current_value:
                    x = neighbor_minus
                    path.append(x)
                    if show_output:
                        print(f"Iteration {iteration + 1}: Moved to x = {x:.4f}, f(x) = {f(x):.4f} (backward)")
                    iteration += 1
                    continue

                break  # No improvement


        if show_output:
            print(f"\nNo improvement found. Terminating at x = {x:.4f}, f(x) = {f(x):.4f}")
            print(f"Path taken: {[f'{xi:.4f}' for xi in path]}")
            print()

        return x, path

    if method == 'simple':
        print("=" * 60)
        print("(a) Simple Hill Climbing")

    return _run_single(x0, method)


if __name__ == "__main__":

    #Simple Hill Climbing
    simple_x, simple_path = hill_climbing(x0=0.0, step_size=0.1, method='simple')
```
